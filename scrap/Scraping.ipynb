# Import Libaries

```python
from selenium import webdriver
import joblib
import numpy as np
import pandas as pd
import csv
from datetime import datetime
import time
import sqlite3
```

# Scrap all cars links

```python
driver = webdriver.Chrome()
links = []
```

```python
for page in range(1, 1452):
    print(f'Scraping page : {page}')
    driver.get(f'https://eg.hatla2ee.com/en/car/page/{page}')
    for header in driver.find_elements_by_class_name('newCarListUnit_header'):
        links.append(header.find_element_by_tag_name('a').get_attribute('href'))
```

```python
driver.close()
```

```python
def get_price():
    try:
        return driver.find_element_by_class_name('usedUnitCarPrice').text
    except:
        return np.nan
```

```python
def get_main_value(car_property, main_data_dict):
    try:
        return main_data_dict[car_property]
    except:
        return np.nan
```
